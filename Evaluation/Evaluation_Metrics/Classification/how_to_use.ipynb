```python
y_test = [0, 1, 1, 0]
y_pred = [0, 1, 0, 0]
y_pred_proba = [0.1, 0.9, 0.4, 0.2]  # probabilities for positive class
```

```python
from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
print("F1 Score:", metrics.f1_score(y_test, y_pred))
print("ROC-AUC:", metrics.roc_auc_score(y_test, y_pred_proba))
```

```python
from sklearn.metrics import (
    accuracy_score as ACC,
    precision_score as PRE,
    recall_score as REC,
    f1_score as F1,
    roc_auc_score as ROC_AUC,
)

print("Accuracy:", ACC(y_test, y_pred))
print("Precision:", PRE(y_test, y_pred))
print("Recall:", REC(y_test, y_pred))
print("F1 Score:", F1(y_test, y_pred))
print("ROC-AUC:", ROC_AUC(y_test, y_pred_proba))
```
