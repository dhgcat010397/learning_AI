# K Nearest Neighbors (KNN)

```python
import matplotlib.pyplot as plt
```

Create synthetic dataset

```python
from sklearn.datasets import make_blobs

# Define parameters for make_blobs
n_samples = 300
centers = [[2, 2], [-2, -2], [2, -2]]
# cluster_std = [0.5, 0.5, 0.5]
cluster_std = [1, 1, 1]

# Create synthetic dataset
X, y = make_blobs(
    n_samples=n_samples, centers=centers, cluster_std=cluster_std, random_state=42
)

# using make_blobs to create a dataset.
# n_samples specifies the total number of points to generate.
# centers specifies the coordinates of the cluster centers.
# cluster_std controls the spread of the clusters around their centers. (standard deviation)
# X contains the feature data (coordinates of points)
# y contains the corresponding cluster labels.
```

```python
# Create the plot
plt.figure(figsize=(8, 6))

# Plot each blob with a different color
colors = ["#FF9999", "#66B2FF", "#99FF99"]
for i in range(3):
    plt.scatter(
        X[y == i, 0],
        X[y == i, 1],
        color=colors[i],
        label=f"Blob {i+1}",
        alpha=0.6,
    )

# Customize the plot
plt.title("Three 2D Gaussian Blobs", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)
```

```python
plt.figure(figsize=(8, 6))

# Plot each blob with a different color
colors = ["#FF9999", "#66B2FF", "#99FF99"]
for i in range(3):
    plt.scatter(
        X_train[y_train == i, 0],
        X_train[y_train == i, 1],
        color=colors[i],
        label=f"Blob {i+1} (Train)",
        alpha=0.6,
    )

for i in range(3):
    plt.scatter(
        X_test[y_test == i, 0],
        X_test[y_test == i, 1],
        marker='*',
        color='#000000',
        s=100,
        label=f"Blob {i+1} (Test)",
        alpha=0.6,
    )

# Customize the plot
plt.title("Three 2D Gaussian Blobs", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Add a slight margin around the blobs
plt.margins(0.1)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Fit and perform K Nearest Neighbors model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN classifier on test set: {accuracy}")
```
