# Decision Tree Algorithm Demonstration

```python
import numpy as np
import matplotlib.pyplot as plt

# using matplotlib.pyplot to create a simple line plot
```

```python
dataset = np.array([
    [1,1,0], [2,1,0], [3,1,0], [4,1,0],
    [1,2,0], [2,2,0], [3,2,0], [4,2,0],
    [1,3,1], [2,3,1], [3,3,0], [4,3,0],
    [1,4,1], [2,4,1], [3,4,0], [4,4,0],
    [1,5,0], [2,5,0], [3,5,0], [4,5,0],
    [1,6,0], [2,6,0], [3,6,0], [4,6,0],
])
```

```python
# Split the dataset into 2 groups
group_a = dataset[dataset[:, 2] == 0]
group_b = dataset[dataset[:, 2] == 1]

# Draw group A with color blue
plt.scatter(group_a[:, 0], group_a[:, 1], c="blue", s=100, label="Group A")

# Draw group B with color red
plt.scatter(group_b[:, 0], group_b[:, 1], c="red", s=100, label="Group B")

# Add labels and legend
plt.xlabel("X-axis: Front legs' length")
plt.ylabel("Y-axis: Back legs' length")
plt.legend()
plt.title("Mouse Dataset Visualization")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# using DecisionTreeClassifier to fit the dataset and plot the decision tree
# using plot_tree to visualize the decision tree
```

```python
# Create features (inputs) and labels (outputs)
X = dataset[:, :2]  # Features: front legs' length and back legs' length
Y = dataset[:, 2]   # Labels: 0 or 1

# Create a decision tree object
decision_tree = DecisionTreeClassifier(criterion="gini")
decision_tree.fit(X, Y)
```

```python
# Plot the tree
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plot_tree(
    decision_tree,
    fontsize=10,
    filled=True,
    feature_names=["Front legs' length", "Back legs' length"],
    class_names=["0", "1"],
)

# Draw group A and B again for reference
plt.subplot(1, 2, 1)
plt.scatter(group_a[:, 0], group_a[:, 1], c="blue", s=100, label="Group A")
plt.scatter(group_b[:, 0], group_b[:, 1], c="red", s=100, label="Group B")

# Show the plot
plt.xlabel("X-axis: Front legs' length")
plt.ylabel("Y-axis: Back legs' length")
plt.legend()
plt.title("Mouse Dataset Visualization")
plt.show()
```
